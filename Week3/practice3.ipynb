```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
import re
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def rotation(img):
    rows, cols = img.shape
    M = cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6)
    image2 = cv.warpAffine(img, M, (cols, rows))

    plotOthers(image2, "Rotated Image")
    return "Process Successful"
```

```python
img = cv.imread('img/modern/unknown.jpg',0)
plotOriginal(img)
rotation(img)
```
