```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
import re
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def reflection(img):
    rows, cols = img.shape
    x = int(input("Enter:\n1 to reflect by row or\n2 for column\n"))
    if(x==1):
        reflect = rows
    else:
        reflect = cols
    M = np.float32([[1,0,0],[0,-1,reflect],[0,0,1]])
    image2 = cv.warpPerspective(img, M, (int(cols), int(rows)))

    plotOthers(image2, "Reflected Image")
    return "Process Successful"
```

```python
img = cv.imread('img/modern/unknown.jpg',0)
plotOriginal(img)
reflection(img)
```
