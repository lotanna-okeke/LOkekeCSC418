```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
import re
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def sheared_Y(img):
    rows, cols = img.shape
    M = np.float32([[1, 0, 0], [0.5, 1, 0], [0,0,1]])
    image2 = cv.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))

    plotOthers(image2, "Sheared in Y-Axis")
    return "Process Successful"
```

```python
img = cv.imread('img/modern/unknown.jpg',0)
plotOriginal(img)
sheared_Y(img)
```
