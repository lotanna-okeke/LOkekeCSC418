```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
import re
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def translation(img):
    rows, cols = img.shape
    x = int(input("Enter the x-axis translation value"))
    y = int(input("Enter the y-axis translation value"))

    M = np.float32([[1,0,x],[0,1,y]])
    image2 = cv.warpAffine(img, M, (cols, rows))

    plotOthers(image2, "Translated Image")
    return "Process Successful"
```

```python
img = cv.imread('img/modern/unknown.jpg',0)
plotOriginal(img)
translation(img)
```
