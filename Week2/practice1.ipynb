```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def visualiseColorSpaces(image):
    channel = int(input('Enter: \n1 to see blue,\n2 to see green, \n3 to see red\n'))
    B,G,R = cv2.split(image)
    if(channel == 1):
        plotOthers(B,"Blue")
    elif(channel == 2):
        plotOthers(G,"Green")
    elif(channel == 3):
        plotOthers(R,"Red")
    else:
        return "Entered invalid value, Process terminated"
    return "Process successful"
```

```python
img = cv2.imread("img/wendy.jpeg")
plotOriginal(img)
visualiseColorSpaces(img)
```
