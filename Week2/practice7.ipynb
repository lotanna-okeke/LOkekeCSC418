```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def inverse(image):
    image2 = 255-image
    plotOthers(image2, "Inverse Color")
    return "Process successful"
    
```

```python
img = cv2.imread("img/wendy.jpeg")
plotOriginal(img)
inverse(img)
```
