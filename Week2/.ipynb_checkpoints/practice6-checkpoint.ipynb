```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
def removeNoise(image):
    image2 = cv2.medianBlur(image, 15)
    plotOthers(image2, "MEDIAN Blur")
    return "Process successful"
    
```

```python
img = cv2.imread("img/wendy.jpeg")
plotOriginal(img)
removeNoise(img)
```
