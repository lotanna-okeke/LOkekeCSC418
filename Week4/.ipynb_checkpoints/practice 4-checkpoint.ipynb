```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the img
image = cv2.imread('img/modern/unknown.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
#resize images
image = cv2.resize(image, (400,400))
```

```python
#Apply laplacian edge detection
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

```python
# convert the output to an 8-bit image
laplacian = cv2.convertScaleAbs(laplacian)
```

```python
plotOriginal(image)
plotOthers(laplacian, "Laplacian Edges")
```
