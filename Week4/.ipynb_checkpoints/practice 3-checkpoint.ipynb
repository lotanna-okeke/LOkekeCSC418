```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the img
image = cv2.imread('img/modern/unknown.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
#perform gausian blur
image = cv2.GaussianBlur(image,(5,5), 0)
```

```python
#define robert kernels
kernel_x = np.array(
    [[1,0],
    [0,-1]])
kernel_y = np.array(
    [[0,1],
    [-1,0]])
```

```python
#apply robert edge detection
robert_x = cv2.filter2D(image, -1, kernel_x)
robert_y = cv2.filter2D(image, -1, kernel_y)
```

```python
#compute magnitude of gradient manually
robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))
```

```python
plotOriginal(image)
plotOthers(robert_combined, "Robert Edges")
```
