```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the img
image = cv2.imread('img/modern/unknown.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def plotOriginal(image):
    #plot the original img
    plt.subplot(1,2,1)
    plt.title("original")
    plt.imshow(image)
```

```python
def plotOthers(image,title):
    plt.subplot(1,2,2)
    plt.title(title)
    plt.imshow(image)
```

```python
#resize images
image = cv2.resize(image, (500,400))
```

```python
#define prewitt kernels
kernel_x = np.array(
    [[-1,-1,-1],
    [0,0,0],
    [1,1,1]])
kernel_y = np.array(
    [[-1,0,1],
    [-1,0,1],
    [-1,0,1]])
```

```python
#apply prewitt edge detection
prewitt_x = cv2.filter2D(image, -1, kernel_x)
prewitt_y = cv2.filter2D(image, -1, kernel_y)
```

```python
#compute magnitude of gradient manually
prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))
```

```python
plotOriginal(image)
plotOthers(prewitt_combined, "Prewittt Edges")
```
